```python
import pandas as pd
import numpy as np
```

### Convert the end-game dataset

```python
df1 = pd.read_csv('c4_game_database.csv')
df1
```

```python
df1.values[:, :-1]
```

```python
# convert to our flattened format
def convert_df1(v):
    # v is flattened 1.0/-1.0 input, output is 1/0 flattened board rep that .reshape(6, 7, 2) would give the correct board
    return np.moveaxis(np.flip(np.hstack([v == 1.0, v == -1]).astype(int).reshape((2, 6, 7)), axis=1), 0, 2).flatten()


df1_ = pd.DataFrame(data={
    'board': df1.apply(lambda x: convert_df1(x.values[:-1]), axis=1),
    'target': df1.apply(lambda x: (x['winner'] + 1) / 2, axis=1)       # make [1, 0, 1] to [0, 0.5, 1]
})
df1_
```

```python
df1_.shape
```

```python
# wins, draws, losses
df1_[df1_['target'] == 1].count()[0], df1_[df1_['target'] == 0.5].count()[0], df1_[df1_['target'] == 0].count()[0]
```

### Convert the 8-ply dataset

```python
df2 = pd.read_csv('connect-4.data')
df2
```

```python
df2.iloc[0, :-1].values.reshape((6, 7), order='F')
```

```python
v = df2.iloc[0, :-1]
z = np.hstack([v == 'x', v == 'o']).astype(int).reshape((6, 7, 2), order='F')
z[:, :, 0] - z[:, :, 1]
```

```python
# convert to our flattened format
def convert_df2(v):
    # v is flattened 1.0/-1.0 input, output is 1/0 flattened board rep that .reshape(6, 7, 2) would give the correct board
    return np.hstack([v == 'x', v == 'o']).astype(int).reshape((6, 7, 2), order='F').flatten()


df2_ = pd.DataFrame(data={
    'board': df2.apply(lambda x: convert_df2(x.values[:-1]), axis=1),
    'target': df2.apply(lambda x: (1.0 if x['win'] == 'win' else (0.5 if x['win'] == 'draw' else 0.0)), axis=1)
})
df2_
```

```python
# wins, draws, losses
df2_[df2_['target'] == 1].count()[0], df2_[df2_['target'] == 0.5].count()[0], df2_[df2_['target'] == 0].count()[0]
```

### Combine 

```python
final_df = pd.concat([df1_, df2_])
final_df
```

```python
final_df.to_hdf('my_c4_dataset.h5', key='c4_dataset', index=False)
```
